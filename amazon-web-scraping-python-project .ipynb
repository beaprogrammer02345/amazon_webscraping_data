```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
URL="https://www.amazon.in/s?k=playstation+4&crid=1PJVS65O790W9&sprefix=playstation+%2Caps%2C423&ref=nb_sb_noss_2"
```

```python
#Header 
HEADERS={"User Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36","Accept-Langauge":"en-US,en;q=0.5"}
```

```python
#webpages
webpages=requests.get(URL,headers=HEADERS)
```

```python
print(webpages)
```

```python
type(webpages)
```

```python
soup=BeautifulSoup(webpages.content,'html.parser')
```

```python
links=soup.find_all("a",attrs={'class': "a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"})
```

```python
links
```

```python
link=links[0].get('href')
```

```python
product_link='https://www.amazon.in/'+link
```

```python
product_link
```

```python
webpages=requests.get(product_link,headers=HEADERS)
```

```python
webpages
```

```python
new_soup=BeautifulSoup(webpages.content,'html.parser')
```
